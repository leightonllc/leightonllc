```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
options = webdriver.EdgeOptions();
options.add_argument('headless');
warnings.filterwarnings("ignore", category=DeprecationWarning)
driver_path = r'C:\Users\User\Documents\edgedriver_win64\msedgedriver.exe'
driver = webdriver.Edge(executable_path=driver_path, options=options)


def mingpao(keyword, times):
    mingpao_link = 'https://news.mingpao.com/php/search2.php?site=mp&pnssection=all&inssection=all&searchtype=A&keywords=' + keyword
    driver.get(mingpao_link)
    driver.implicitly_wait(15)
    i = 0
    for article in driver.find_elements_by_class_name('color6th'):
        i += 1
        if i == times + 1:
            break
        try:
            header = article.find_element(By.CLASS_NAME, 'highlightsearchword')
            print(article.find_element(By.TAG_NAME, 'p').text)
            print(header.find_element(By.TAG_NAME, 'a').text)
            print(header.find_element(By.TAG_NAME, 'a').get_attribute('href'))
            print()
        except:
            pass
        
def scmp(keyword, times):
    scmp_link = 'https://www.scmp.com/search/' + keyword
    driver.get(scmp_link)
    driver.implicitly_wait(15)
    i = 0
    for article in driver.find_elements_by_class_name('search-results__item'):
        i += 1
        if i == times + 1:
            break
        try:
            print(article.find_element(By.CLASS_NAME, 'wrapper__published-date').text)
            print(article.find_element(By.CLASS_NAME, 'search-page__title-container').text)
            print(article.find_element(By.TAG_NAME, 'a').get_attribute('href'))
            print()
        except:
            pass

def financial_times(keyword, times):
    ft_link = 'https://www.ft.com/search?q=' + keyword + '&sort=date'
    driver.get(ft_link)
    driver.implicitly_wait(15)
    i = 0
    for article in driver.find_elements_by_class_name('search-results__list-item'):
        i += 1
        if i == times + 1:
            break
        try:
            print(article.find_element(By.CLASS_NAME, 'o-teaser__timestamp').text)
            header = article.find_element(By.CLASS_NAME, 'o-teaser__heading')
            print(header.text)
            print(header.find_element(By.TAG_NAME, 'a').get_attribute('href'))
            print()
        except:
            pass

        
mingpao('ESG', 7)
#mingpao('Green+Finance', 7)
#mingpao('Climate', 3)
#financial_times('ESG', 7)
#financial_times('Green+Finance', 7)
#financial_times('Climate', 7)
#scmp('ESG', 7)
#scmp('Green+Finance', 7)
#scmp('Climate', 7)
```
